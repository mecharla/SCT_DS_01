Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reads the CSV file into a Pandas DataFrame

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

Display the first few rows of the dataset to understand what data is available
(Shows the top 5 rows of the dataset)


```python
print(df.head())
```

Visualize distribution of a categorical column (e.g., 'Sex' or 'Pclass')

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Sex', data=df)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

 Visualize distribution of a continuous column (e.g., 'Age')

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
